```python
# Databricks notebook source
# =====================================================
# PREPARACION DE AMBIENTE
# Proyecto: ETL Hydro - Arquitectura Medallion
# Crea:
#   - Catalog: catalog_final_project
#   - Schemas: bronze, silver, gold
# Idempotente (no falla si ya existen)
# =====================================================

CATALOGO = "catalog_final_project"
SCHEMA_BRONZE = "bronze"
SCHEMA_SILVER = "silver"
SCHEMA_GOLD   = "gold"

print("=== PREPARACION DE AMBIENTE ===")
print("Catalogo:", CATALOGO)

# COMMAND ----------

# 1️⃣ Crear catálogo (Unity Catalog)
spark.sql(f"CREATE CATALOG IF NOT EXISTS {CATALOGO}")

# Asegurar uso del catálogo
spark.sql(f"USE CATALOG {CATALOGO}")

print("Catalog validado.")

# COMMAND ----------

# 2️⃣ Crear schemas (capas Medallion)
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOGO}.{SCHEMA_BRONZE}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOGO}.{SCHEMA_SILVER}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOGO}.{SCHEMA_GOLD}")

print("Schemas creados / validados:")
print(f"- {SCHEMA_BRONZE}")
print(f"- {SCHEMA_SILVER}")
print(f"- {SCHEMA_GOLD}")

# COMMAND ----------

# 3️⃣ Validación final (útil para logs CI/CD)
display(spark.sql(f"SHOW SCHEMAS IN {CATALOGO}"))

print("=== PREP_AMB COMPLETADO ===")
```
